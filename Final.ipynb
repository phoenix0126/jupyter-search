```python
import requests
from IPython.display import display, Markdown, clear_output
r = requests.get("http://makeup-api.herokuapp.com/api/v1/products.json")
makeup_data = r.json()
makeup_data
```

```python
types= {}
typelist = []
for element in makeup_data:
    if element['product_type'] in types:
        types[element['product_type']] += 1
    else: 
        types[element['product_type']] =1
        typelist = typelist + [element['product_type']]
types
```

```python
prices= {}
pricelist = []
for element in makeup_data:
    if element['price'] != None:
        if element['price'] in prices:
            prices[element['price']] += 1
        else:
            prices[element['price']] =1
            pricelist = pricelist + [float(element['price'])]
prices
```

```python
brands= {}
brandlist = []
for element in makeup_data:
    if element['brand'] != None:
        if element['brand'] in brands:
            brands[element['brand']] += 1
        else: 
            brands[element['brand']] =1
            brandlist = brandlist + [element['brand']]
brands
```

```python
import ipywidgets as widgets
from IPython.display import HTML, display
import pandas as pd

productType_radio = widgets.RadioButtons(
    options=typelist,
    description='Product Type:',
    disabled=False
)

price_slider = widgets.IntRangeSlider(
    value=[min(pricelist), max(pricelist)],
    min=min(pricelist),
    max=max(pricelist),
    step=1,
    description='Price:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)

brand_select = widgets.SelectMultiple(
    options = brandlist,
    value=[],
    rows=10,
    description='Brands:',
    disabled=False
)

box = widgets.VBox([productType_radio, price_slider, brand_select])
box

filterbutton = widgets.Button(description='Filter')
out = widgets.Output()

def filter_product(_):
    with out:
        clear_output()
        filteredproducts = []
        for element in makeup_data:
            if ( element['price'] != None and element['brand'] != None and element['product_type'] == productType_radio.value and float(element['price']) >= price_slider.value[0] and float(element['price']) <= price_slider.value[1] and ( element['brand'] in brand_select.value or brand_select.value == () ) ):
                filteredproduct = {}
                filteredproduct['name'] = element['name']
                filteredproduct['image_link'] = element['image_link']
                filteredproducts.append(filteredproduct)
                
        pd.set_option("display.max_colwidth", None)
        table = pd.DataFrame(filteredproducts)
        display(table)
        
filterbutton.on_click(filter_product)
widgets.VBox([filterbutton,out])
```
